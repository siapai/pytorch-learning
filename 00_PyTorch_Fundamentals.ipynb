### Matrix Multiplication

Two main ways to perform multiplication

1. Element-wise Multiplication
2. Matrix Multiplication (dot product)

There are two main rules that performing matrix multiplication needs to satisfy:
1. The **inner dimensions** must match
* `(3, 2) @ (3, 2)` wont work
* `(2, 3) @ (3, 2)` will work
*`(3, 2) @ (2, 3)` will work

2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# Element-wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

```python
%%time
torch.matmul(tensor, tensor)
```

## Reshaping, stacking, squeezing, and unsqueezing tensors

* Reshaping - Reshapes an input tensor to a defined shaped
* View - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* Stacking - Combine multiple tensors on top of each other (Vstack) or side by side (Hstack)
* Squeeze - Remove all `1` diemnsions from a tensor
* Unsqueeze - Add `1` dimension to target tensor
* Permute - Return a view of the input dimensions permuted (swapped) in a certain way

```python
# Let's create a tensor
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add an extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
# Change the view
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes x (beause a view of a tensor share same memory)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=0)
x_stacked, x_stacked.shape
```

```python
# Squeeze tensor
print(f"Previous tensor: {x_reshaped}")
print(f"Previous shape: {x_reshaped.shape}")

# Remove extra dimensions from x_reshaped
x_squeezed = x_reshaped.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"\nNew shape: {x_squeezed.shape}")
```

```python
# Unsqueeze tensor
print(f"Previous tensor: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add extra dimensions with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"\nNew shape: {x_unsqueezed.shape}")
```

```python
# Permute tensor - Rearrange in to spesific order
x_original = torch.rand(size=(224, 224, 3))

# Permute the original tensor to reaarange the axis (or dim) order
x_permuted = x_original.permute(2,  0, 1)

print(f"Prev shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

```python
x_original[0, 0, 0] = 77777
x_original[0, 0, 0], x_permuted[0, 0, 0]
```

## PyTorch tensors and NumPy

NumPy is a popular scientific python numerical computing library

And because of this, PyTorch has functionality to interact with it

* Data in NumPy, want on PyTorch tensor -> `torch.from_numy(ndarray)`
* PyTorch tensor -> NumPy -> `torch.Tensor.numpy()`

```python
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array).type(torch.float32)
array, tensor
```

```python
# Change the value of array, what will this do to `Tensor`?

array = array + 1
array, tensor
```

```python
# Tensor to NumPy array
tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Change the tensor, what happens to `numpy_tensor`
tensor = tensor + 1
tensor, numpy_tensor
```

## Reproducibility (trying to take random out of random)

In short how NN learns:

`Start with random numbers -> tensor operation -> update random numbers to try and make them better representations of the data -> again -> again...`

To reduce the randomness in NN and PyTorch comes the concept of a **random seed**

Essenstially what the random seed does is "flavour" the randomness

```python
# Create two random tensors
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)
print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# Let's make some random but reporucible tensors

RANDOM_SEED = 42

torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
